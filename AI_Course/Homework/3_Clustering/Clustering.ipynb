## Clustering

### KMeans

```python
def permi(nn,begin,end):
    global COUNT
    if begin>=end:
        rst_lst.append(dict(zip([0,1,2,3,4,5],nn)))
        COUNT +=1
    else:
        i=begin
        for num in range(begin,end):
            nn[num],nn[i]=nn[i],nn[num]
            permi(n,begin+1,end)
            nn[num],nn[i]=nn[i],nn[num]
COUNT=0

rst_lst=[]

n=[0,1,2,3,4,5]

permi(n,0,len(n))
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

random_state = 200 #随机数种子

# Incorrect number of clusters
y_pred = KMeans(n_clusters=6, random_state=random_state).fit_predict(x)

temp_count =0
temp_index=0
index=0

for dic in rst_lst:
    mapped=[dic[i] if i in dic else i for i in y_pred]

    count=0
    for i in range(len(mapped)):
        if mapped[i]==data['expression'][i]:
            count+=1;
    if count>temp_count:
        temp_count=count
        index=temp_index
    temp_index+=1
    
print(temp_count/len(mapped))
```

### Mean Shift

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
bandwidth = estimate_bandwidth(x, quantile=0.06, n_samples=50)

ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit_predict(x)
```
